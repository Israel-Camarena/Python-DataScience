```python
texto = input(" Introduce la frase que deseas analizar: ")
```

```python
caracteres = len(texto)
caracteres_sin_espacios = len(texto.replace(" ", ""))
vocales = sum(texto.lower().count(v) for v in "aeiou")
palabras = len(texto.split())
Eliminar_Palabra = texto.strip().split(" ", 1)[-1]
Espacio_Giones = texto.replace(" ", "-")
MinuscalasAMayusculas = texto.upper()
MayusculasAMinusculas = texto.lower()
```

```python
print(f"Numeros de caracteres: {caracteres}" 
      f"\nNumeros de caracteres sin espacios: {caracteres_sin_espacios}" 
      f"\nNumeros de vocales: {vocales}" 
      f"\nNumeros de palabras: {palabras}" 
      f"\nTexto con la primera palabra eliminada: {Eliminar_Palabra}" 
      f"\nEspacio por guiones: {Espacio_Giones}" 
      f"\nMinusculas a mayusculas: {MinuscalasAMayusculas}" 
      f"\nMayusculas a minusculas: {MayusculasAMinusculas}")
```

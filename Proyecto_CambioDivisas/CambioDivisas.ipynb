```python
nombre = input("Bienvenido a la calculadora de cambio de divisas, porfavor ingresa tu nombre: ")

# Obtener la hora actual
from datetime import datetime
fecha = datetime.now() 
Hora = fecha.hour # obtener la hora actual

# Definir el tiempo mediante la hora
# Mañana: 5:00 - 11:59
# Tarde: 12:00 - 17:559
# Noche: 18:00 - 04:59
if 5 <= Hora < 12:
    print("Buenos días, " + nombre + "!, empezamos con la calculadora de cambio de divisas.")
elif 12 <= Hora < 18:
    print("Buenas tardes, " + nombre + "!, empezamos con la calculadora de cambio de divisas.")
else:
    print("Buenas noches, " + nombre + "!, empezamos con la calculadora de cambio de divisas.")

#Solicitar al usuario que ingrese la cantidad de dinero a convertir
cantidad = float(input("Porfavor" + nombre + ", ingresa la cantidad de dinero que deseas convertir: "))
```

```python
# Definir las tasas de cambio de divisas
tasas_camibio = {
    "USD": 0.058 , #Dolar estadunidense
    "EUR": 0.049, #Euro
}

# calcular el cambio de divisas
def convertir_divisas(cantidad, tasas_camibio):
    cantidad_usd = cantidad * tasas_camibio["USD"]
    cantidad_eur = cantidad * tasas_camibio["EUR"]
    return cantidad_usd, cantidad_eur

# Obtener el resultado de la conversion
print("El cambio de divisas es el siguiente: ")
cantidad_usd, cantidad_eur = convertir_divisas(cantidad, tasas_camibio)
print("Dolar estadunidense (USD): " + str(cantidad_usd))
print("Euro (EUR): " + str(cantidad_eur))
```

```python
# cambio de billetes y monedas solo acepta billetes de 100, 50, 20 y monedas de 10,5,1
def cambio_billetes_monedas_usd(cantidad_usd):
    billetes_100 = cantidad_usd // 100
    billetes1 = (cantidad_usd - (billetes_100 * 100))//10
    billetes_50 = cantidad_usd // 50
    billetes2 = (cantidad_usd - (billetes_50 * 50))//5
    billetes_20 = cantidad_usd // 20
    billetes3 = (cantidad_usd - (billetes_20 * 20))//2
    monedas_10 = cantidad_usd % 10
    monedas_5 = cantidad_usd % 5
    monedas_1 = cantidad_usd % 1
    return billetes1, billetes2, billetes3, monedas_10, monedas_5, monedas_1

# Obtener el resultado del cambio de billetes y monedas
print("El cambio de billetes y monedas es el siguiente: ")

billetes1, billetes2, billetes3, monedas_10, monedas_5, monedas_1 = cambio_billetes_monedas_usd(cantidad_usd)
print("Billetes de 100: " + str(billetes1))
print("Billetes de 50: " + str(billetes2))
print("Billetes de 20: " + str(billetes3))
print("Monedas de 10: " + str(monedas_10))
print("Monedas de 5: " + str(monedas_5))
print("Monedas de 1: " + str(monedas_1))
```

```python
# Desarrolla un programa que le pregunte al usuario la cantidad de kilómetros que desea convertir a millas.

# Utiliza la función input() para recoger esta información, conviértela a un valor numérico y realiza la conversión (considera que 1 kilómetro equivale a 0.621371 millas). Finalmente, imprime el resultado.
# Instrucciones:
# 1. Al iniciar el programa, debes pedir al usuario que ingrese una cantidad de kilómetros utilizando la orden siguiente: "Introduce la cantidad de Kilómetros recorridos: "  y almacenar esta entrada en una variable llamada km

# 2. Luego, convierte los kilómetros a millas considerando que 1 kilómetro equivale a 0.621371 millas  y guarda el resultado.

# 3. Finalmente, imprime el resultado de esta operación.
```

```python
km = float(input("Introduce la cantidad de Kilómetros recorridos: "))
millas = km * 0.621371
print(millas)
```
